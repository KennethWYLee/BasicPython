```python
dataset = pd.read_excel("AAPL.xlsx", sheet_name="sheet1")
```

```python
dataset[:200]
```

```python
dataset[200:]
```

```python
train_set = dataset[:200]
test_set = dataset[200:]
```

```python
train_y = train_set["y"]
train_y
```

```python
train_x = train_set[["x1", "x2", "x3"]]
train_x
```

```python
test_y = test_set["y"]
test_y
```

```python
test_x = test_set[["x1", "x2", "x3"]]
test_x
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(train_x, train_y)

reg.coef_
```

```python
# reg.predict(train_x)
```

```python
y_train_hat = reg.predict(train_x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(train_y, y_train_hat)
```

```python
y_test_hat = reg.predict(test_x)
```

```python
mean_squared_error(test_y, y_test_hat)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg_decision = DecisionTreeRegressor()
```

```python
reg_decision.fit(train_x, train_y)
```

```python
y_train_hat_decision = reg_decision.predict(train_x)
```

```python
mean_squared_error(train_y, y_train_hat_decision)
```

```python
y_test_hat_decision = reg_decision.predict(test_x)
```

```python
mean_squared_error(test_y, y_test_hat_decision)
```
